```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
# from sklearn.linear_model import LogisticRegression
# from sklearn.svm import SVC
# from sklearn.tree import DecisionTreeClassifier
# from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv('breast_cancer_data.csv')
df.head()
```

```python
#Handling missing values
df=df.drop(columns={'Unnamed: 32'})
df.head()
```

```python
#label encoding
encode=LabelEncoder()
labels=encode.fit_transform(df['diagnosis'])
df['target']=labels
```

```python
df.head()
```

##M--1 D--0

```python
x=df.drop(columns={'diagnosis','target'})
y=df['target']
```

```python
print(x)
print(y)
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=2,random_state=2)
print(x_train.shape,x_test.shape)
print(x_train)
```

```python
#Standardization
standard=StandardScaler()
x_train_std=standard.fit_transform(x_train)
x_test_std=standard.fit(x_test)
```

```python
print(x_train_std)
```

```python
#model evaluation
model = KNeighborsClassifier(n_neighbors=10)  # You can change K value
model.fit(x_train_std, y_train)
```

```python
# Standardization
standard = StandardScaler()
x_train_std = standard.fit_transform(x_train)
x_test_std = standard.transform(x_test)

# Train the model
model = KNeighborsClassifier(n_neighbors=10)
model.fit(x_train_std, y_train)

# Predict and evaluate
y_predict = model.predict(x_test_std)
score = accuracy_score(y_test, y_predict) 
print("Accuracy Score:", score * 100)
```

```python
hyperparamtuning = GridSearchCV(model, param_grid={
    'n_neighbors': [2, 3, 5, 10, 50, 70,1]
}, cv=5)

hyperparamtuning.fit(x_train, y_train)

print("Best K value:", hyperparamtuning.best_params_)
print("Best score:", hyperparamtuning.best_score_)
```
